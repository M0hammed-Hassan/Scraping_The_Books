```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="http://books.toscrape.com/catalogue/category/books/science_22/index.html" #This our web site link.
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,"html.parser")
```

```python
def get_title(): #This function include all titles of books.
    title=[]
    for i in soup.find_all("h3"):
        title.append(i.get_text())
    return title
def get_price(): #This function include the price for each book.
    price=[]
    lst=[]
    for i in soup.find_all("p",attrs={'class':'price_color'}):
        lst.append(i.get_text())
    for i in lst:
        price.append(str(i)[1:]) #For select only usefl chars.
    return price
def get_rate(): #This function include all the rates for all books.
    lst1=[]
    lst2=[]
    lst3=[]
    rate=[]
    for i in soup.find_all("p"):
        lst1.append(str(i)[9:27]) #I selected this range for making sure i took all the desired chars.
    for i in range(0,42,3): #Now i'm appending only the rate.
        lst2.append(lst1[i]) 
    for i in lst2: #For deleting > from some rates.
        if str(i).endswith('>'):
            lst3.append(str(i)[0:17])
        else:
            lst3.append(i)
    for i in lst3: #For deleting " from the rates. 
        rate.append(str(i)[1:len(i)-1])
    return rate
```

```python
df=pd.DataFrame(data={'Title':get_title(),'Price':get_price(),'Rate':get_rate()}) #Print data as data frame.
```
